```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

def run_stage2_gridsearch(X_train, y_train, preprocessor):
    results = []

    # ----------------------------
    # Model definitions + param grids
    # ----------------------------
    model_dict = {
        "Decision Tree": (
            DecisionTreeClassifier(random_state=42),
            {
                'model__max_depth': [6, 8, 10, 12],
                'model__min_samples_split': [10, 15, 20],
                'model__min_samples_leaf': [1, 2, 4]
            }
        ),
        "SVC": (
            SVC(probability=True, class_weight='balanced'),
            {
                'model__C': [1.0, 1.5, 2.0, 3.0],
                'model__gamma': ['scale', 0.01, 0.05, 0.1, 0.2],
                'model__kernel': ['rbf']
            }
        ),
        "Random Forest": (
            RandomForestClassifier(random_state=42),
            {
                'model__n_estimators': [400, 600, 800, 1000],
                'model__max_depth': [8, 10, 12, 15],
                'model__min_samples_split': [10, 15, 20],
                'model__min_samples_leaf': [1, 2, 4]
            }
        ),
        "XGBoost": (
            XGBClassifier(random_state=42, eval_metric='logloss'),
            {
                'model__n_estimators': [500, 700, 900],
                'model__max_depth': [3, 4, 5],
                'model__learning_rate': [0.03, 0.04, 0.05],
                'model__subsample': [0.75, 0.85, 0.95],
                'model__colsample_bytree': [0.75, 0.85, 1.0],
                'model__gamma': [0, 0.1, 0.2],
                'model__reg_alpha': [0, 0.3],
                'model__reg_lambda': [1, 1.2]
            }
        )
    }

    cv = 5

    # ----------------------------
    # Run GridSearchCV for each model
    # ----------------------------
    for name, (estimator, param_grid) in model_dict.items():
        pipe = Pipeline([
            ('preprocess', preprocessor),
            ('model', estimator)
        ])
        grid = GridSearchCV(pipe, param_grid, cv=cv, scoring='accuracy', n_jobs=-1)
        print(f"Running GridSearchCV for {name}...")
        grid.fit(X_train, y_train)
        
        results.append({
            'Model': name,
            'Best CV Accuracy': grid.best_score_,
            'Best Parameters': grid.best_params_
        })

    # ----------------------------
    # Convert results to DataFrame and sort
    # ----------------------------
    results_df = pd.DataFrame(results).sort_values(by='Best CV Accuracy', ascending=False).reset_index(drop=True)
    return results_df

```
